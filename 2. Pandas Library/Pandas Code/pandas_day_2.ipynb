# How to Rename Column

```python
df = pd.DataFrame(np.random.rand(4,8), columns=list("ABCDEFGH"))
df
new_column_names = ["New_Column1", "New_Column2", "New_Column3", "New_Column4", "New_Column5", "New_Column6", "New_Column7", "New_Column8"]
df.columns = new_column_names
df
```

# Another way to rename columns names

```python
df
# Define a dictionary that maps current column names to new column names
column_mapping = {
    "New_Column1": 'samad',
    "New_Column2": 'samad',
    "New_Column3": 'samad',
    "New_Column4": 'samad',
    "New_Column5": 'samad',
    "New_Column6": 'samad',
    "New_Column7": 'samad',
    "New_Column8": 'samad',
}
# Use the rename() method to rename the columns
df.rename(columns=column_mapping, inplace=True)
df
```

```python
# chaning only one column name
df.rename(columns={"samad": "Random"}, inplace=True)
df
```

```python
# chaning any character and string inside column
column_names = ['First_', 'Second_', 'Third_', 'Fourth_4', 'Fifth_']
my_df = pd.DataFrame(np.random.rand(3,5), columns=column_names)
my_df

# now lets change the _ to *
my_df.columns = my_df.columns.str.replace("_", "*")
my_df
```

# Adding prefix or suffix to columns

```python
my_df = my_df.add_prefix("S_")
my_df = my_df.add_suffix("woow")
my_df
```
