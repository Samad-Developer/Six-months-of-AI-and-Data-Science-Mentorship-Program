```python
df.shape
df.sample(frac=0.2).shape
```

n Parameter (Number of Samples):

n: Number of random samples to select from the DataFrame. E.g., df.sample(n=5) selects 5 random rows.
frac Parameter (Fraction of Samples):

frac: Fraction of the DataFrame to sample, ranging from 0 to 1. E.g., df.sample(frac=0.2) selects 20% of the rows.

axis Parameter (Sampling Columns Instead of Rows):

axis=1: Sample columns instead of rows. E.g., df.sample(n=2, axis=1) selects 2 random columns.
