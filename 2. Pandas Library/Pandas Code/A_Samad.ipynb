Read the csv file to load data into df

```python
import pandas as pd
df = pd.read_csv('./datasets/googleplaystore.csv')
df.sample(2)
```

1.  Replace "Varies with device" with NaN

```python
df['Size'] = df['Size'].str.replace('Varies with device', 'NaN')
df.sample(2)
```

2. Convert size from KB to MB and remove m and k etc.

```python
def convert_kb_to_mb(size):
    try:
        # Remove non-numeric characters and convert to float
        size = float(''.join(filter(str.isdigit, str(size))))
    except (ValueError, TypeError):
        # Handle errors by returning a default value (e.g., NaN)
        return None  # You can use None, np.nan, or any other value as needed
    if 'k' in str(size):
        return size / 1024
    else:
        return size

# Apply the updated function to the 'Size' column
df['Size'] = df['Size'].apply(convert_kb_to_mb)
```
