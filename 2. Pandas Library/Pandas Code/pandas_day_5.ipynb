```python
import pandas as pd
import seaborn as sns
data = {'Name': ['John Smith', 'Alice Johnson', 'Bob Williams', 'Emily Davis'],
        'Location': ['New York, USA', 'Los Angeles, USA', 'London, UK', 'Sydney, Australia']}
my_data = pd.DataFrame(data)
my_data
```

# Splliting string into separate columns

```python
my_data[["firstName", "lastName"]] = my_data["Name"].str.split(" ", expand=True)
my_data
```

```python
my_data[['city', 'country']] = my_data['Location'].str.split(', ', expand=True)
my_data
```

```python
my_data = my_data[['firstName', 'lastName', 'city', 'country']]
my_data
```
