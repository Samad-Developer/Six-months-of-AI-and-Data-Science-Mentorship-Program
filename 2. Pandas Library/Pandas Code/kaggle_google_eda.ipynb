```python
data = pd.DataFrame({
    'A': [10, 20, 30, 40, 50,4],
    'B': [5, 15, 85, 35, 45,8]
})

# Find the index label of the maximum value in column 'A'
index_of_max_A = data['A'].idxmax()

# Find the index label of the maximum value in the entire DataFrame
index_of_max_all = data.idxmax().max()

print("Index label of the maximum value in column 'A':", index_of_max_A)
print("Index label of the maximum value in the entire DataFrame:", index_of_max_all)
```

```python
def stars(row):
    if row.Category == 'FAMILY':
        return 4
    elif row.Rating >= 4:
        return 4
    else:
        return 0
start_rating = df.apply(stars, axis='columns')
start_rating
```
