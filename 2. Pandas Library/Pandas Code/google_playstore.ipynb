```python
# conver the whole size column into bytes

# let's define a function

def convert_size(size):
    if isinstance(size, str):
        if 'k' in size:
            return float(size.replace('k', '')) * 1024
        elif 'M' in size:
            return float(size.replace('M', '')) * 1024 * 1024
        elif 'Varies with device' in size:
            return np.nan
    return size
```

```python
df['Size']
```

```python
# let's apply the function
df['Size'] = df['Size'].apply(convert_size)
df['Size']
```

```python
df.head()
```

```python
# rename the column
df.rename(columns={"Size": 'Size_in_bytes'}, inplace=True)
df.head()
```

```python
# now convert size_in_byte to size in MB
df['Size_in_MB'] = df['Size_in_bytes'].apply(lambda sizeBytes: sizeBytes/(1024 * 1024))
df.head()
```

# Installs Column:
1. Remove + sign
2. remove comma ,
3. convvert the column into an integer

```python
df['Installs'] = df['Installs'].apply(lambda installs: installs.replace('+','') if '+' in str(installs) else installs)
```

```python
df['Installs'].value_counts()
```

```python
df['Installs'] = df['Installs'].apply(lambda installs: installs.replace(',','') if ',' in str(installs) else installs)
```

```python
df['Installs'].value_counts()
```

```python
df['Installs'] = df['Installs'].apply(lambda installs: int(installs))
```

```python
df['Installs'].dtype
```

# Plot Diagram for missing values

```python
plt.figure(figsize=(16,6))
missing_percentage = (df.isnull().sum())/(len(df)*100)
missing_percentage.plot(kind='bar')

plt.xlabel('columns')
plt.ylabel('percentage')
plt.title('Percentage of Missing Values in each column')
```

```python
missing_percentage[missing_percentage < 1].plot(kind='bar')
```
