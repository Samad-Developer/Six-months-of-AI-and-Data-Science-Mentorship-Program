# Linear Algebra in python

```python
import numpy as np
import matplotlib.pyplot as plt
```

Equation is <br>
2x + 3y = 8 <br>
5x + 4y = 20

```python
# coefficient matrix A
A = np.array([[2, 3], [5, 4]])
A
```

```python
# constant matrix B
B = np.array([8, 20])
B
```

```python
# solve for x
x = np.linalg.solve(A, B)
print("The solution is: ", x)
```
