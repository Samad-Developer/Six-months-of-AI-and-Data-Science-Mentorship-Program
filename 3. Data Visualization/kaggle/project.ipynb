```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
my_data = pd.read_csv('../../2. Pandas Library/datasets/dc-wikia-data.csv')
```

```python
my_data.head(10)
```

```python
yearly_appearances = my_data.groupby('YEAR')['APPEARANCES'].sum()
yearly_appearances
```

```python
# Plot a line chart to show the trend in character appearances over the years
yearly_appearances.plot(kind='line', figsize=(12, 6))
plt.title('Comic Character Appearances Over Time')
plt.xlabel('Year')
plt.ylabel('Total Appearances')
plt.show()
```
