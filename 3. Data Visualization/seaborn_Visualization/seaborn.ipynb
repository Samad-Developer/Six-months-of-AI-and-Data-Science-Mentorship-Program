```python
g = sns.FacetGrid(data=tips, col='sex', row='time')
g.map_dataframe(sns.scatterplot, x='total_bill', y='tip')
g.set_axis_labels("Total Bill Amount", "Tip Amount")
g.add_legend()
g.set_titles(col_template="{col_name} customers")
```
