# Data Visualization

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import titanic data from seaborn library
titanic_data = sns.load_dataset('titanic')
```

# We will use pandas for visualization

```python
titanic_data.plot(kind='scatter', x='age', y='fare', color='blue')
```

```python
titanic_data.groupby('sex')['age'].mean().plot(kind='bar')
plt.xlabel('sex')
plt.ylabel('Age')
```

```python
# seaborn plots
# bar plot using seaborn on titanic_dataset

sns.barplot(data=titanic_data, x='sex', y='age', hue='class')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

```python
sns.boxplot(data=titanic_data, x='sex', y='age', hue='class')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

```python
# make scatter plot via plotly
import plotly.express as px

# Assuming you have a DataFrame called 'titanic_data' with x and y columns
fig = px.box(titanic_data, x='age', y='fare', color='class')

fig.show()
```

```python
conditions = ['Group A', 'Group B', 'Group C']
heights = [15, 20, 18]
std_dev = [1.5, 2.0, 1.8]

# Create a bar chart with error bars
plt.bar(conditions, heights, yerr=std_dev, capsize=5)

# Add labels and title
plt.xlabel('Plant Groups')
plt.ylabel('Average Height (cm)')
plt.title('Plant Growth Experiment Results')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import axes3d

plt.style.use('_mpl-gallery')

# Make data
X, Y, Z = axes3d.get_test_data(0.05)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
