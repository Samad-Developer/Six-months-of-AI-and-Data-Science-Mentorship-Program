# Line plot in pandas

Line plots are commonly used in data visualization when you want to represent data points as connected line segments. They are suitable for displaying data that has a natural ordering, such as time series data or data where there's a continuous relationship between the variables. Here are some common scenarios where line plots are appropriate:

1. **Time Series Data**: Line plots are excellent for showing how a variable changes over time. For example, you can use line plots to visualize stock prices over a period, temperature variations throughout the day, or website traffic over months.

2. **Trends and Patterns**: Line plots help reveal trends, patterns, and fluctuations in data. They are useful for identifying seasonal patterns, long-term trends, and short-term variations.

3. **Comparing Multiple Series**: You can create line plots with multiple lines to compare different variables or categories. For instance, you can compare the sales performance of multiple products or the revenue of different departments.

4. **Interpolation**: Line plots can be used to interpolate values between known data points. This is particularly useful when you have data at discrete time intervals, and you want to estimate values at non-sample times.

5. **Tracking Changes Over Categories**: Line plots can show how a variable changes over categories. For instance, you can visualize how the performance of various sales regions changes over time.

6. **Showing Continuous Data Relationships**: Line plots are well-suited for demonstrating continuous relationships between variables. You can use them to visualize the relationship between variables like age and income, showing how income changes with age.

7. **Correlation Analysis**: When you want to visualize the relationship between two variables, a line plot can help identify positive or negative correlations.

8. **Projecting Future Trends**: Line plots can be used to make projections about future trends based on historical data. This is common in financial forecasting and economic analysis.

It's important to note that line plots are most effective when the data has a natural order or progression. If the data points are not connected in a meaningful way, other types of plots, like scatter plots or bar plots, might be more appropriate. Ultimately, the choice of plot depends on the data and the story you want to tell.

# Basic line plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Temperature_Celsius': [5, 8, 12, 18, 22, 26],
    'Precipitation_mm': [50, 45, 60, 30, 20, 15]
}

df = pd.DataFrame(data)

# Create a line plot with various customizations

# Customize the line style, color, and marker
df.plot(x='Month', y='Temperature_Celsius', kind='line', linestyle='-', color='green', marker='o')

# Set labels for the x and y axes
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')

# Set a title for the plot
plt.title('Monthly Temperature Trends')

# Show a grid on the plot
plt.grid(True)

# Add a legend
plt.legend(['Temperature (°C)'])

# Adjust the figure size
plt.figure(figsize=(10, 6))

# Customize x-axis tick positions and labels
# xticks = [1, 2, 3, 4, 5, 6]
# xtick_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
# plt.xticks(xticks, xtick_labels)

# Set y-axis limits
# plt.ylim(0, 30)

```

## Multiple line plot

```python

# Sample data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'New York': [32, 36, 45, 57, 68, 76, 82, 80, 73, 61, 49, 37],
    'Los Angeles': [65, 68, 70, 73, 76, 79, 82, 84, 83, 79, 72, 67],
    'Chicago': [30, 34, 45, 58, 68, 77, 82, 80, 73, 61, 48, 35]
}

df = pd.DataFrame(data)
df.set_index('Month', inplace=True)

# Create a multiple line plot
df.plot(kind='line', title='Monthly Temperature Trends')

# Customize the plot
plt.xlabel('Month')
plt.ylabel('Temperature (°F)')
plt.legend(title='Cities', loc='upper left')
plt.grid(True)

plt.show()
```
