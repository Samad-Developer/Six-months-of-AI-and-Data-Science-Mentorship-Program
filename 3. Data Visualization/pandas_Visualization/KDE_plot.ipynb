# Kernel Density Estimation (KDE) or density plots

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = [45, 62, 57, 50, 71, 65, 54, 48, 63, 58, 55, 59, 62, 47, 52, 64, 59, 49, 58, 53, 56]

# Create a pandas Series
series = pd.Series(data)

# Create a KDE plot
series.plot.kde(title='Kernel Density Estimation Plot', color='skyblue')

# Customize the plot
plt.xlabel('Values')
plt.ylabel('Density')

plt.show()
```
