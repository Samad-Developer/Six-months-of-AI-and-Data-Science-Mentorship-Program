# Hist plot

A histogram (hist plot) is used to visualize the distribution of a single continuous variable or numeric data. It shows the frequency of data points within specified intervals, allowing you to understand the underlying data distribution. Histograms are commonly used in data analysis and statistics to reveal patterns and characteristics of data. Here's an example of creating a histogram using pandas:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data of exam scores
exam_scores = [78, 85, 92, 65, 88, 72, 95, 80, 91, 75, 84, 89, 68, 78, 82, 93, 87, 76, 90, 81]

# Create a DataFrame
df = pd.DataFrame(exam_scores, columns=['Scores'])

# Create a histogram
df['Scores'].plot(kind='hist', bins=6, rwidth=0.85, edgecolor='black')

# Customize the plot
plt.xlabel('Score Ranges')
plt.ylabel('Number of Students')
plt.title('Distribution of Exam Scores')

plt.show()
```
