# Box Plot

A box plot (box-and-whisker plot) is used to display the distribution and spread of a dataset, particularly its five-number summary: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. Box plots are especially useful for identifying outliers and comparing the distribution of multiple datasets. Here's a meaningful example of creating a box plot using pandas:

Suppose we have exam scores for two different classes, Class A and Class B, and we want to compare their score distributions using box plots.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for exam scores in two classes
data = {
    'Class A': [78, 85, 92, 65, 88, 72, 95, 80, 91, 75, 84, 89, 68, 78, 82, 93, 87, 76, 90, 81],
    'Class B': [68, 73, 82, 57, 78, 64, 91, 76, 89, 72, 75, 88, 62, 69, 79, 85, 74, 77, 87, 80]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a box plot to compare the two classes
df.plot(kind='box', title='Box Plot of Exam Scores for Two Classes')

# Customize the plot
plt.ylabel('Scores')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
